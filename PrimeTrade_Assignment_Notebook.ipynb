## Trader Performance vs Bitcoin Fear & Greed Sentiment

```python
import pandas as pd
import matplotlib.pyplot as plt
hist=pd.read_csv('historical_data.csv')
fg=pd.read_csv('fear_greed_index.csv')
```

```python
hist['date']=pd.to_datetime(hist['Timestamp'],unit='ms').dt.date
fg['date']=pd.to_datetime(fg['date']).dt.date
df=hist.merge(fg[['date','classification']],on='date',how='left')
df.head()
```

```python
analysis=df.dropna(subset=['classification'])
summary=analysis.groupby('classification')['Closed PnL'].agg(['count','mean','sum'])
print(summary)
summary['mean'].plot(kind='bar')
plt.title('Average PnL by Sentiment')
plt.show()
```
